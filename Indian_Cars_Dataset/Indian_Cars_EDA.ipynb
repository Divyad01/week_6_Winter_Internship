# Different car manufacturers  using pie charts.

```python
# Counting the occurrences of each car manufacturer
make_counts = df['make'].value_counts()

# Plotting  a pie chart for the contribution of different car manufacturers
plt.figure(figsize=(10, 8))

# Customizing the layout parameters to improve readability
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85,
        wedgeprops=dict(width=0.4), textprops=dict(fontsize=10))

plt.title('Contribution of Car Manufacturers')

# Adjusting the layout for better spacing
plt.tight_layout()

plt.show()
```

# Count plots for categorical variables(navigation_system and child_safety_locks)

```python
# Categorical columns for count plots
categorical_columns = ['child_safety_locks', 'navigation_system']

# Setting the style of seaborn
sns.set(style="whitegrid")

# Creating count plots for each categorical column
plt.figure(figsize=(10, 5))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(1, 2, i)
    sns.countplot(x=column, data=df, palette='viridis')
    plt.title(f'Count Plot of {column}')

# Adjusting layout for better spacing
plt.tight_layout()

plt.show()
```

# Bivariate Analysis

## Correlation Matrix:


```python
# Select numeric columns for correlation matrix
numeric_columns = ['price', 'cylinders', 'fuel_tank_capacity', 'city_mileage', 'seating_capacity', 'basic_warranty', 'number_of_airbags']

# Create a correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

## Scatter Plots 

```python
# Select numeric columns for scatter plot
numeric_columns = ['price', 'city_mileage']

# Set the style of seaborn
sns.set(style="whitegrid")

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='city_mileage', data=df, color='skyblue', edgecolor='black')
plt.title('Scatter Plot: Price vs City Mileage')
plt.xlabel('Price')
plt.ylabel('City Mileage')
plt.show()
```

## Pair Plot

```python
# Selecting numeric columns for pair plot
numeric_columns = ['price', 'cylinders', 'fuel_tank_capacity', 'city_mileage', 'seating_capacity', 'basic_warranty', 'number_of_airbags']

# Setting the style of seaborn
sns.set(style="ticks")

# Creating a pair plot
sns.pairplot(df[numeric_columns])
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()
```

## Bar Charts with Hue

```python
# Categorical columns for bar chart
categorical_column = 'make'
additional_categorical_column = 'body_type'

# Settiing the style of seaborn
sns.set(style="whitegrid")

# Creating a bar chart with hue
plt.figure(figsize=(12, 6))
sns.countplot(x=categorical_column, hue=additional_categorical_column, data=df, palette='viridis')
plt.title(f'Count of {categorical_column} with Hue {additional_categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Count')
plt.legend(title=additional_categorical_column, loc='upper right')
plt.show()
```

## Heatmaps:

```python
# Select numeric columns for correlation matrix
numeric_columns = ['price', 'cylinders', 'fuel_tank_capacity', 'city_mileage', 'seating_capacity', 'basic_warranty', 'number_of_airbags']

# Create a correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

plt.title('Correlation Matrix Heatmap of Numeric Variables')
plt.show()
```

# Joint Plots

```python
numeric_columns = ['price', 'city_mileage']

# Setting the style of seaborn
sns.set(style="whitegrid")

# Creating a joint plot
sns.jointplot(x='price', y='city_mileage', data=df, kind='scatter', color='skyblue')

plt.suptitle('Joint Plot of Price and City Mileage', y=1.02)
plt.show()
```

## Box Plots with Hue 

```python
# Categorical and numeric columns for box plot
categorical_column = 'make'
numeric_column = 'price'
additional_categorical_column = 'body_type'

# Setting the style of seaborn
sns.set(style="whitegrid")

# Creating a box plot with hue
plt.figure(figsize=(12, 8))
sns.boxplot(x=categorical_column, y=numeric_column, hue=additional_categorical_column, data=df, palette='viridis')

plt.title(f'Box Plot of {numeric_column} with Hue {additional_categorical_column} by {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel(numeric_column)
plt.legend(title=additional_categorical_column, loc='upper right')
plt.show()
```

## Violin plot

```python
# Create a violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x=categorical_column, y=numeric_column, data=df, palette='viridis')

plt.title(f'Violin Plot of {numeric_column} by {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel(numeric_column)
plt.show()
```

# Stacked Bar Charts:


```python
#Categorical columns for stacked bar chart
category_column = 'make'
sub_category_column = 'body_type'
numeric_column = 'price'

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a stacked bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=category_column, y=numeric_column, hue=sub_category_column, data=df, palette='viridis', ci=None)

plt.title(f'Stacked Bar Chart of {numeric_column} by {category_column} and {sub_category_column}')
plt.xlabel(category_column)
plt.ylabel(numeric_column)
plt.legend(title=sub_category_column, loc='upper right')
plt.show()
```
