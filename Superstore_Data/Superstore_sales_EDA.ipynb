# Calculating the quantity

```python
# Group by product-related columns and aggregate Quantity
product_quantity = df.groupby(['Product ID', 'Product Name'])['Sales'].count().reset_index()
product_quantity.rename(columns={'Sales': 'Quantity'}, inplace=True)

# Merge the product_quantity DataFrame back to the original DataFrame
df = pd.merge(df, product_quantity, on=['Product ID', 'Product Name'], how='left')

# Check the updated DataFrame with the 'Quantity' column
print(df[['Product Name', 'Sales', 'Quantity']])
```

# Time Series Plots: Plot time series graphs to understand the trends and patterns in sales over the 4-year period.


```python
#Converting the 'Order Date' column to a datetime format.
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year  # Extracting only the year

# Group by year and sum the sales
sales_by_year = df.groupby('Year')['Sales'].sum()

# Plot time series graph for sales by year
plt.figure(figsize=(12, 6))
plt.plot(sales_by_year.index, sales_by_year.values, color='blue', marker='o', linestyle='-')
plt.title('Time Series Plot of Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

#  Seasonal Decomposition: Decompose time series data into components like trend, seasonality, and residuals for deeper insights.
 

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decomposing time series data
result = seasonal_decompose(df['Sales'], model='additive', period=12) 

# Plotting the decomposed components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(result.observed, label='Observed')
plt.legend()
plt.title('Observed')

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.legend()
plt.title('Trend')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.legend()
plt.title('Seasonal')

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residuals')
plt.legend()
plt.title('Residuals')

plt.tight_layout()
plt.show()
```

# Sales Variation Over Time: Plot line charts to observe how sales vary over different time periods (months, quarters, years).


## Monthly Sales Variation:

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)

# Resampling data to get monthly sales
monthly_sales = df['Sales'].resample('M').sum()

# Creating line chart for monthly sales variation
plt.figure(figsize=(12, 6))
monthly_sales.plot(marker='o', color='skyblue')
plt.title('Monthly Sales Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Monthly Sales')
plt.show()
```

## Quarterly Sales Variation:

```python
# Resampling data to get quarterly sales
quarterly_sales = df['Sales'].resample('Q').sum()

# Creating line chart for quarterly sales variation
plt.figure(figsize=(12, 6))
quarterly_sales.plot(marker='o', color='lightcoral')
plt.title('Quarterly Sales Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Quarterly Sales')
plt.show()
```

## Yearly Sales Variation:

```python
# Resampling data to get yearly sales
yearly_sales = df['Sales'].resample('Y').sum()

# Create line chart for yearly sales variation
plt.figure(figsize=(12, 6))
yearly_sales.plot(marker='o', color='lightgreen')
plt.title('Yearly Sales Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Yearly Sales')
plt.show()
```

# BiVariate Analysis

## Correlation Analysis: Examine the correlation between sales and other relevant numeric variables.


```python
numeric_columns = ['Sales', 'Postal Code']  # Add other numeric columns as needed

# Selecting relevant columns for correlation analysis
correlation_data = df[numeric_columns]

# Calculating the correlation matrix
correlation_matrix = correlation_data.corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## Scatter Plots: Plot scatter plots to explore the relationship between sales and another numeric variable.

## Pair Plots: Use pair plots for a quick overview of relationships between multiple numeric variables.


```python

new_df = df.groupby(df.index).agg({'Sales': 'sum', 'Quantity': 'sum', 'Postal Code': 'first'})

# Select relevant numeric columns for pair plot
numeric_columns = ['Sales', 'Quantity']

# Create pair plot
sns.pairplot(new_df[numeric_columns])
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()
```

## Discount Impact on Sales: Analyze the impact of discounts on sales using bar charts or line charts.

```python
df['Discount'] = pd.cut(df['Sales'], bins=[0, 100, 200, 300, 400, 500], labels=['0-100', '101-200', '201-300', '301-400', '401-500'])

# Group by 'Discount' and calculate average sales for each discount level
discount_sales = df.groupby('Discount')['Sales'].mean().reset_index()

# Create a bar chart or line chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Discount', y='Sales', data=discount_sales, palette='viridis')


plt.title('Impact of Discounts on Sales')
plt.xlabel('Sales Range')
plt.ylabel('Average Sales')
plt.show()
```

## Product-wise Sales Analysis: Investigate how sales vary for different products or categories using bar charts or line charts.
 

```python
# Bar chart for total sales per product
plt.figure(figsize=(15, 6))
product_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).reset_index()
sns.barplot(x='Sales', y='Product Name', data=product_sales.head(10), palette='viridis')

plt.title('Top 10 Products by Sales')
plt.xlabel('Sales')
plt.ylabel('Product Name')
plt.show()
```

## Region-wise Product Sales: Compare product sales across different regions using grouped bar charts.
 

```python
# Group by 'Region' and 'Product Name', then sum the sales
region_product_sales = df.groupby(['Region', 'Product Name'])['Sales'].sum().reset_index()

# Creating a grouped bar chart
plt.figure(figsize=(15, 8))
sns.barplot(x='Product Name', y='Sales', hue='Region', data=region_product_sales.head(10), palette='viridis')

plt.title('Top 10 Products Sales Comparison Across Regions')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
